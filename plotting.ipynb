# Practice

```python
with open('student_data.txt','r') as f:
    student_yr=[]
    student_id=[]
    data=f.readlines()
    #Loop through line in the data set
    for line in data[1:]:
        line=line.strip() #Remove \n

        #Splitted each line into indivdule values
        split_line=line.split()
        #Added values to their lists.
        student_id.append(split_line[0])
        student_yr.append(split_line[1])

    # print(len(student_id))
    # print(len(student_yr))

    plt.scatter(student_id, student_yr)
    plt.show()
```

# Problem 1

```python
with open('time_and_temp.txt','r') as f:
    time=[]
    temp=[]
    data=f.readlines()
    for line in data[1:]:
        striped_line=line.strip()
        split_line=striped_line.split()
        time.append(float(split_line[0]))
        temp.append(float(split_line[1]))


plt.scatter(time,temp,color='green',marker='o',linestyle='-')
plt.title('Time and Temperature')
plt.xlabel('Time (secs)')
plt.ylabel('Temperature (c)')
plt.show()

        
```

# Problem 2

```python
with open('temp_and_pressure.txt','r') as f:
    pressure=[]
    temp=[]
    data=f.readlines()
    for line in data[1:]:
        striped_line=line.strip()
        split_line=striped_line.split()
        pressure.append(float(split_line[0]))
        temp.append(float(split_line[1]))

plt.scatter(pressure, temp, color='red', marker='o')
plt.title('Temperature and Pressure')
plt.xlabel('Temperature (C)')
plt.ylabel('Pressure (lbs per square inch)')
plt.show()
```
